```python
import pandas as pd
```

```python
api_key = "apikey"
api_secret = "apisecret" #api key and secret key from binance to authorize trading
```

```python
client = Client(api_key, api_secret) #setting up client with keys
```

```python
pd.DataFrame(client.get_historical_klines('BTCUSDT', '1m', '30 min ago UTC') )
#gets bitcoin data from 30 minutes ago in 1 minute intervals
```

```python
def cleanData(coin, interval, lookback): #function to clean the data and return clean datafram of wanted information
    frame = pd.DataFrame(client.get_historical_klines(coin, interval, lookback + ' min ago UTC')) #creates dataframe
    frame = frame.iloc[:, :6] #we only need first 6 columns of data for this project
    frame.columns = ['Time', 'Open', 'High', 'Low', 'Close', 'Volume'] #these are given in the website information
    frame = frame.set_index('Time') #time is set as index
    frame.index = pd.to_datetime(frame.index, unit = 'ms') #time is by default in unix, so we convert to easier time
    frame = frame.astype(float)
    return frame
```

```python
test = cleanData('DOGEUSDT', '1m', '30')
```

```python
#buy is asset fell by more than 0.3% within last 30 min
#sell if asset rises by more than 0.2% or falls further than 0.15% within last 30 min
def basicStrat(coin, quantity, bought = False):
    df = cleanData(coin, '1m', '30m') #creates dataframe
    cumulReturn = (df.Open.pct_change() + 1).cumprod() - 1 #calculates cumulative return
    if not bought:
        if cumulReturn[-1] < -0.003: #if coin drops by 0.3%, then buy
            order = client.create_order(coin = coin, side = 'BUY', type = 'MARKET', quantity = quantity)
            print(order)
            bought = True
        else:
            print('No Trade Executed')
    if bought:
        while True:
            df = cleanData(coin, '1m', '30m') #updates dataframe
            sinceBuy = df.loc[df.index > pd.to_datetime(order['transactionTime'], unit = 'ms')] #checks time since purchase
            if len(sinceBuy) > 0:
                sinceBuyRet = (sinceBuy.Open.pct_change() + 1).cumprod() - 1 #calculates cumulative return
                if sinceBuyRet[-1] > 0.002 or sinceBuyRet[-1] < -0.0015: #if one of the two are fulfilled then sell
                    order = client.create_order(coin = coin, side = 'SELL', type = 'MARKET', quantity = quantity)
                    print(order)
                    break
            
```
